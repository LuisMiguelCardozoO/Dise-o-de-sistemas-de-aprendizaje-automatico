```python
import numpy as np
import matplotlib.pyplot as plt
import math


import pandas as pd
datos = pd.read_csv('diabetes2.csv')
##datos

datos[['BMI','Outcome']].head()

#datos[['BMI','Outcome']].plot.scatter(x='BMI',y='Outcome')
#datos[['Glucose','Outcome']].plot.scatter(x='Glucose',y='Outcome')

from sklearn.linear_model import LogisticRegression

# definiendo input y output
X_train = np.array(datos['BMI']).reshape((-1, 1))
Y_train = np.array(datos['Outcome'])

# creando modelo
model = LogisticRegression()
model.fit(X_train, Y_train)

# imprimiendo parametros
print(f"intercepto (b): {model.intercept_}")
print(f"pendiente (w): {model.coef_}")


```

```python
w = 0.09
b = -3.6

# puntos de la recta
x = np.linspace(0,datos['BMI'].max(),100)
y = 1/(1+np.exp(-(w*x+b)))

# grafica de la recta
datos.plot.scatter(x='BMI',y='Outcome')
plt.plot(x, y, '-r')
plt.ylim(0,datos['Outcome'].max()*1.1)
plt.title("Función logística Estándar", fontsize=14.0)

plt.show()

```

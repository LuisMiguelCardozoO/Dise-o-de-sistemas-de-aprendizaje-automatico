```python
import pandas as pd
datos = pd.read_csv('train.csv')
datos
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 7))
plt.ylabel("SalePrice", fontsize=20)
plt.xlabel("GrLivArea", fontsize=20)
plt.scatter(datos["GrLivArea"], datos["SalePrice"], color="pink", s=50, label='data')
plt.legend(loc='upper left', fontsize=20)
plt.show()
```

```python
from sklearn import linear_model

regresion = linear_model.LinearRegression()
GrLivArea = datos["GrLivArea"].values.reshape((-1,1))
modelo = regresion.fit(GrLivArea,datos["SalePrice"])
print("IntersecciÃ³n (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
entrada = [[200],[500],[912],[1200],[3000],[5000]]
modelo.predict(entrada)

fig = plt.figure(figsize=(12, 7))
plt.scatter(entrada, modelo.predict(entrada), color="red", s=50)
plt.plot(entrada, modelo.predict(entrada), color="black" )

plt.ylabel("SalePrice", fontsize=20)
plt.xlabel("GrLivArea", fontsize=20)
plt.scatter(datos["GrLivArea"], datos["SalePrice"], color="pink", s=50, label='data')
plt.legend(loc='upper left', fontsize=20)
plt.show()
```
